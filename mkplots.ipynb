```python
from test import main

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fwk_hist, fwk_test_acc, pth_hist, pth_test_acc = main(num_training_runs=10)
```

```python
''' Plot the loss and accuracy from the time series generated by main()
    requires seaborne to plot the standard deviation in a nice way
'''


def plotTrain (history, title="") : 
    
    f, fig1 = plt.subplots(figsize=(6,4))
    
    df = pd.concat([pd.DataFrame(history[i]) for i in range(len(history))])
    df["Epoch"] = df.index
    
    sns.lineplot(x='Epoch', y='train_loss', data=df, label='training loss', ax=fig1)
    
    fig1.yaxis.grid(True)
    fig1.set_title(title)
    fig1.set_xlabel('Epoch')
    fig1.set_ylabel('Loss')
    
    
def plotVal (history, title="") : 
    
    f, fig1 = plt.subplots(figsize=(6,4))
    
    df = pd.concat([pd.DataFrame(history[i]) for i in range(len(history))])
    df["Epoch"] = df.index
    
    #$sns.lineplot(x='Epoch', y='train_loss', data=df, label='training loss', ax=fig1)
    sns.lineplot(x='Epoch', y='val_loss', data=df, label='validation loss', ax=fig1)
    
    acc_scale_multiplier = plt.ylim()[1]
    print ("acc_scale_multiplier : ", acc_scale_multiplier)
    df["val_acc"] = df["val_acc"] * acc_scale_multiplier # hack, but necessary
    sns.lineplot(x='Epoch', y='val_acc', data=df, label='validation accuracy', ax=fig1)
    
    fig1.yaxis.grid(True)
    fig1.set_title(title)
    fig1.set_xlabel('Epoch')
    fig1.set_ylabel('Loss')
    
    fig1_y2 = fig1.twinx()
    fig1_y2.set_ylabel('Accuracy')
    
```

```python
plotTrain (fwk_hist, title="Our Framework : training loss")
plotVal (fwk_hist, title="Our Framework : validation loss & validation accuracy")
```

```python
plotTrain (pth_hist, title="Pytorch : training loss")
plotVal (pth_hist, title="Pytorch : validation loss & validation accuracy")
```
